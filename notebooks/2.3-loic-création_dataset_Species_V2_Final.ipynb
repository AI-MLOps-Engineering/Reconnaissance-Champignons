```python
'''
Script permettant de classer les images dans les sous-rГ©pertoires relatifs Г  leur 'species', pour les champignons du Top 30 en France.

Cela permettra d'utiliser l'utilitaire 'image_dataset_from_directory' pour appeler les images dans les futurs modГЁles et gagner en
rapiditГ© d'exГ©cution.

+ AJout d'un filtre Г  partir de ResNet50 (ImageNet) pour retirer les images parasites

'''
```

```python
# import des librairies nГ©cessaires

import os
import shutil
import pandas as pd
import numpy as np


from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image


from sklearn.utils import resample # pour les mГ©thodes de rГ©chantillonnage (over-sampling et under-sampling)
```

```python
# DГ©finition des chemins d'accГЁs

rep_main_csv = 'C:\\Users\\Utilisateur\\Documents\\DataScience\\Mushroom\\' # Chemin vers le fichier csv principal (Г  part car trop gros pour mettre sur git)
rep_data = "C:\\Users\\Utilisateur\\Documents\\DataScience\\avr25cds_reconnaissance_champignons\\datasets\\" # Chemin vers le fichier CSV 'champignons_france_top30.csv et le dataset "final dataset_30_species.csv"
rep_img_src = 'C:\\Users\\utilisateur\\Documents\\DataScience_images_Source\\' # Chemin vers l'ensemble des images de champignons 
rep_img = 'C:\\Users\\Utilisateur\\Documents\\DataScience_images_Especes\\' # Chemin vers l'ensemble des images de champignons CLASSEES
```

```python
# Lecture du dataset 'observations_mushroom.csv'
mushroom = pd.read_csv(rep_main_csv+'observations_mushroom.csv', sep =  ",")

# Lecture du dataset 'top_30_France.csv'
top_30_France = pd.read_csv(rep_data+'champignons_france_top30.csv', sep = ';', encoding='latin-1')

# Fusion des 2 Datasets
mushroom = mushroom.merge(right = top_30_France, left_on='gbif_info/species', right_on='Nom scientifique', how = 'inner')

# Filtre pour un indice de confiance >= 92
mushroom = mushroom.loc[mushroom['gbif_info/confidence'] >= 92]

mushroom.to_csv(rep_data+"dataset_30_species.csv", index=False, encoding="utf-8")

# Affichage extrait du dataset et de ses caractГ©ristiques
display(mushroom.head())
display(mushroom.info())
```

```python
# Simplification du fichier
mushroom = mushroom[['image_lien', 'gbif_info/species', 'Nom commun', 'Statut', 'Habitat typique']]
mushroom = mushroom.dropna()
mushroom = mushroom.rename({'image_lien' : 'image', 'gbif_info/species' : 'target'}, axis = 1)

# Affichage extrait du dataset simplifiГ© et de ses caractГ©ristiques
display(mushroom.head())
display(mushroom.info())
```

```python
# DГ©finition des catГ©gories qui nous intГ©ressent
categories = mushroom['target'].unique()

```

```python
# Nombre d'images par espГЁces avant over/under-sampling :

print("Nombre d'images par espГЁce avant over/under-sampling :\n")
for category in categories:
    order_list = mushroom.loc[mushroom['target']==category]
    count = len(order_list)
    print(f"L'espГЁce {category} compte {count} enregistrements")

```

PRE-TRAITEMENT IMAGE_NET POUR RETIRER LES IMAGES PARASITES POTENTIELLES

```python
# Charger le modГЁle ResNet50 prГ©-entraГ®nГ© sur ImageNet
model = ResNet50(weights="imagenet")

# Liste des classes "champignons" dans ImageNet
fungi_classes = {
    "bolete", "agaric", "gyromitra", "stinkhorn",
    "earthstar", "hen-of-the-woods", "coral fungus", "mushroom"
}

def champi(img_path, top=3): # on check si il y a champignon dans le top 3 des recos
    """
    VГ©rifie si une image contient un champignon
    Args:
        img_path: chemin vers l'image
        top: nombre de prГ©dictions Г  analyser
    Returns:
        True si l'image est identifiГ©e comme champignon
    """
    # Charger et prГ©traiter l'image
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)  # permet d'avoir x.shape = (1, 224, 224, 3) compatible avec le modГЁle
    x = preprocess_input(x)

    # PrГ©dictions
    preds = model.predict(x)
    decoded = decode_predictions(preds, top=top)[0]

    # VГ©rifie si une des prГ©dictions est un champignon
    return any(label in fungi_classes for _, label, _ in decoded)



```

```python
for i, row in mushroom.iterrows():
    image_path = os.path.join(rep_img_src, row['image'])

    if champi(image_path):
        mushroom.loc[i, ["Champi"]] = True
    else :
        mushroom.loc[i, ["Champi"]] = False
```

```python
display(mushroom.head())
```

```python
# Filtre : on retire les images parasites
mushroom = mushroom.loc[mushroom['Champi'] == True]
```

```python
display(mushroom.info())
```

```python
# Nombre d'images par espГЁces aprГЁs prГ©-filtre :

print("Nombre d'images par espГЁce avant over/under-sampling :\n")
for category in categories:
    order_list = mushroom.loc[mushroom['target']==category]
    count = len(order_list)
    print(f"L'espГЁce {category} compte {count} enregistrements")
```

```python
# ParamГ©trage de l'under-sampling / under-sampling

# Objectif : rГ©duire Г  1000 images les catГ©gories sur-reprГ©sentГ©es et augmenter Г  500 images les catГ©gories sous-reprГ©sentГ©es


# random_state :
rs = 42

# Liste pour stocker les DataFrames Г©quilibrГ©s
list_temp = []

for category in categories:
    df_category = mushroom[mushroom['target']==category].copy()
    size = len(df_category)
    
    # Under-sampling
    if size > 1000:
        df_temp = resample(
            df_category,
            replace = False,
            n_samples = 1000,
            random_state = rs
        )
    
    # Over-sampling
    elif size < 500:
        df_temp = resample(
            df_category,
            replace = True,
            n_samples = 500,
            random_state = rs
        ).reset_index(drop=True)


    # RAS si 500 <= size <= 1000
    else :
        df_temp = df_category
       

    list_temp.append(df_temp)


# Fusion de tous les DataFrames
mushroom_2 = pd.concat(list_temp, ignore_index=True)


# VГ©rification
print(mushroom_2['target'].value_counts())
print("Taille totale :", len(mushroom_2))
```

```python
print(mushroom_2['image'].value_counts())
```

```python
# CrГ©ation de la colonne image sans-doublon "image_unique" avec incrГ©mentation, qui permettra de dupliquer les photos

dico = {}
new = []
for val in mushroom_2['image']:
    if val not in dico:
        dico[val] = 0
        new.append(val)
    else:
        dico[val] += 1
        new.append(f"{val.replace('.jpg', '')}_{dico[val]}.jpg")


mushroom_2['image_unique'] = new

display(mushroom_2.tail())
```

```python
# VГ©rification
print(mushroom_2['image_unique'].describe())
```

COPIE DES IMAGES

```python
# CrГ©ation des sous-dossiers pour chaque catГ©gorie

for category in categories:
    category_path = os.path.join(rep_img, category)
    os.makedirs(category_path, exist_ok=True)
```

```python
# DГ©placement des fichiers dans leur dossier de catГ©gorie

for _, row in mushroom_2.iterrows():
    src_path = os.path.join(rep_img_src, row['image'])
    dst_path = os.path.join(rep_img, row['target'], row['image_unique'])

    # VГ©rification que le fichier existe avant de dГ©placer
    if os.path.exists(src_path):
        shutil.copy(src_path, dst_path)
    else:
        print(f"Fichier introuvable : {src_path}")
    
```

```python
# Nombre final dвЂ™images par catГ©gories

print("Nombre d'images par espГЁce aprГЁs over/under-sampling :\n")
for category in categories:
    final_count = len(os.listdir(rep_img+category))
    print(f"CatГ©gorie {category} Г©quilibrГ©e Г  {final_count} images")
```
